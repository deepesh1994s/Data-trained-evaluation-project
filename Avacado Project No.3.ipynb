# finding the best random_state

assigning x and y variable

```python
x=hf.iloc[:,0:-1]
x.shape
```

```python
y=hf.iloc[:,-1]
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    LR=LogisticRegression()
    LR.fit(x_train,y_train)
    predrf=LR.predict(x_test)
    acc=accuracy_score(y_test,predrf)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("best accuracy is",maxAccu,"on random_state",maxRS)
```

# Creating train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=98)
```

# Model building

```python
LR=LogisticRegression()
LR.fit(x_train,y_train)
predlr=LR.predict(x_test)
print(accuracy_score(y_test,predlr))
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

```python
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)
preddt=DTC.predict(x_test)
print(accuracy_score(y_test,preddt))
print(confusion_matrix(y_test,preddt))
print(classification_report(y_test,preddt))
```

```python
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
predrf=RF.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
AD=AdaBoostClassifier()
AD.fit(x_train,y_train)
predad=AD.predict(x_test)
print(accuracy_score(y_test,predad))
print(confusion_matrix(y_test,predad))
print(classification_report(y_test,predad))
```

# Cross Validation Score

```python
scr=cross_val_score(LR,x,y,cv=5)
print('cross validation score of logisticregression is :',scr.mean())
print(scr.std())
```

```python
scr=cross_val_score(DTC,x,y,cv=5)
print('cross validation score of Decisiontreeclassifier is :',scr.mean())
print(scr.std())
```

```python
scr=cross_val_score(RF,x,y,cv=5)
print('cross validation score of RandomForestclassifier is :',scr.mean())
print(scr.std())
```

```python
scr=cross_val_score(AD,x,y,cv=5)
print('cross validation score of AdaBoostclassifier is :',scr.mean())
print(scr.std())
```
