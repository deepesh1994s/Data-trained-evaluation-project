# Train Test Split

```python
x=df.iloc[:,0:-1]
x.head()
```

```python
y=df.iloc[:,-1]
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# MODEL BUILDING

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
LR=LogisticRegression()
LR.fit(x_train,y_train)
predlr=LR.predict(x_test)
print(accuracy_score(y_test,predlr))
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

```python
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
predrf=RF.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)
preddtc=DTC.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

here we can see that highest accuracy is shown with RandomForestClassifier but it due to be of Overfitting so no we go for cross validation score

# Cross Validation Score

```python
from sklearn.model_selection import cross_val_score
```

```python
scr=cross_val_score(LR,x,y,cv=5)
print('cross validation score of logisticregression is :',scr.mean())
print(scr.std())
```

```python
scr=cross_val_score(DTC,x,y,cv=5)
print('cross validation score of Decisiontreeclassifier is :',scr.mean())
print(scr.std())
```

```python
scr=cross_val_score(RF,x,y,cv=5)
print('cross validation score of RandomForestclassifier is :',scr.mean())
print(scr.std())
```

we can see here DecisionTreeClassifier has the best cross validation score(81%) with minimum std of(.0025) so we go furhter process with dtc

# Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
parameters = {'splitter':["best","random"],'max_depth': np.arange(2,15),'criterion':['gini','entropy'],'max_features':['auto','sqrt','log2']}
```

```python
GCV=GridSearchCV(DecisionTreeClassifier(),parameters,cv=5)
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python
GCV_pred=GCV.best_estimator_.predict(x_test)
accuracy_score(y_test,GCV_pred)
```

by hyper parameter tuning our accuracy score is approx.83%
